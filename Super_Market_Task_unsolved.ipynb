# Q7
* ## Calculate the number of shipping days between the "Order Date" and "Ship Date" columns
* ## And calculate the maximum shipping days and the city of it 

```python
data.columns
```

```python
data['Order Date']=pd.to_datetime(
    data["Order Date"] ,
    dayfirst=True
)

data['Ship Date'] =pd.to_datetime(
    data['Ship Date'] , 
    dayfirst=True
)
```

```python
data['Shipping Days'] = (data['Ship Date'] - data['Order Date']).dt.days
```

```python
max_days = data['Shipping Days'].max()

max_days
```

```python
city=data.loc[data['Shipping Days'].idxmax(),'City']

city
```
